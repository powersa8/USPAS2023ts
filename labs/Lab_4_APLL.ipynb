# Analog Phase Locked Loop

## 1. Introduction

The analog phase-locked loop (APLL) is a control system that generates an output signal whose phase is locked to the phase of an input signal. A PLL can also track an input frequency, generating a frequency that is a multiple of the input frequency. Phase-locked loops are widely employed in radio, telecommunications, computers and other electronic applications.

A PLL is usually made of a signal generator, a phase detector, a loop filter and a voltage-controlled oscillator. Figure 1 shows the schematic of a phase-locked loop



Fig. 1 APLL diagram

The **signal generator** provides a RF reference signal for the whole PLL system. The **voltage-controlled oscillator (VCO)** generates a periodic signal. 
The **phase detector** compares the phase of the VCO signal with the phase of the reference signal. 
The phase error between the two signals will drive the VCO and adjust the oscillator to keep the phase matched. 

When the two signal inputs are equal in phase and frequency, the error will be constant and the loop is locked. 
Bringing the output signal back toward the input signal for comparison is called a feedback loop since the output is fed back toward the input forming a loop. 
The PLL implements negative feedback here.

**Signal Generator** 

We generate a 100 MHz RF reference signal (+6dBm), because the center frequency of VCO is 100MHz.
    
**Voltage-Controlled Oscillator** 

A voltage-controlled oscillator or VCO is an electronic oscillator whose oscillation frequency is controlled by a voltage input. The applied input voltage determines the instantaneous oscillation frequency. The VCXO (voltage controlled crystal oscillator) used in this experiment is one kind of VCOs. A VCXO oscillates with a frequency determined by a crystal, but which can be adjusted over a narrow range with a control voltage, usually in the range of several volts. 

There are many parameters to characterize a VCXO. The most important ones are the tuning voltage range, center frequency, and pulling (tuning) range. The **tuning voltage range** is the variation range of the VCXO control voltage. The **center frequency** is the frequency in the middle of the VCXO’s output frequency range. The **pulling range** is the ratio of the up (or down) frequency variation to the center frequency. This ratio is usually represented in ppm (part per million). Normally, the pulling range is about 100ppm - 200ppm, depending on the structure of VCXO and the selected crystal. Usually a VCXO is followed by a low pass filter which is to suppress harmonics and get a purified signal.
    
**Phase Detector** 

In an analog PLL, a frequency mixer acts as a phase detector. In electronics, a frequency mixer is a nonlinear electrical circuit that creates new frequencies (the sum and difference of the original frequencies) from two signals applied to it. Actually, the frequency mixer here can be modeled as a signal multiplier.

**Loop Filter** 

A low pass filter (LPF) is implemented here to reject the frequency sum signal from the phase detector and pass the frequency difference signal. A first-order RC LPF works well in this PLL system. The cut-off frequency of the LPF is about `tens of kHz` in this experiment.

**Amplifier** 

The value of the error signal from the loop filter should cover the tuning voltage range of VCXO, so we need an amplifier here. We can simply make an amplifier with an operational amplifier (OP-AMP) and several resistors. Figure 3 below is a diagram of the loop filter and amplifier for reference.
