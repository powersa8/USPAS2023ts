## Goal: Demonstrate repetition rate control of modelocked oscillator, locked to stable RF clock. 

**Key elements:**

Modelocked laser with cavity length control, phase locked loop, RF frequency reference**

**What is happening:**

A modelocked laser has voltage-controlled cavity length, making it like a VCO (voltage-controlled oscillator) when RF is derived from the pulse train. Thus, it can be controlled in a phase-locked loop (PLL) when compared with a stable RF reference, and be locked to a distributed clock. In this experiment, the output pulse train is detected by a high speed photodiode, and the fundamental and a high harmonic are derived from the electronic pulse train by filters. These two frequencies are compared on a mixer with corresponding frequencies from the RF reference, and used to control the modelocked laser timing. Since the laser can be locked to a higher frequency, this allows for finer temporal control with less noise, but it creates an ambiguity (bucket ambiguity) because any cycle of the high harmonic can be locked to, not just the one that corresponds with a zero crossing of the fundamental. To find the right bucket, both frequencies would have to be transmitted and processed.

## 2. PIEZO AND MOTOR TUNING RANGE, OPEN LOOP. 

With the loop unlocked (PID amp in manual), and the mixer **IF** output displayed on the slow scope, at a sweep speed sufficient to get several cycles on the screen, measure the frequency. 
This is the beat between the (assumed constant) reference RF and the detected laser RF. 

Vary the piezo amp offset voltage using the front panel knob, observing the voltage on the front panel meter. 
Go from zero to 150V. 

Write down the beat frequencies at those voltages. 
There may be slow drift of the laser frequency, especially since the piezo is being exercised. (If the frequency range goes through zero, this could be confusing, so if thats the case, offset things just a tad by tapping the stepping motor momentary, bidirectional toggle switch. Dont hold it down or up, just pulse it as short as you can.) 

Given that the laser repetition rate is 100MHz, and knowing that the signal at the mixer is the 28th harmonic, how far would the piezo have to move (and given that its at the end of the cavity where the effective length change is double-passed) to make the frequency change you see? 
This is how the laser works as a voltage-controlled oscillator (VCO). 
The piezo excursion will be on the order of 10 microns.

## 3. OPERATION OF THE PLL. 

Before locking the PLL again, return the piezo amp to 75V offset, so it can operate near the middle of the range. 

Make the beat frequency as low as possible by using the stepping motor, bringing the frequency within the range of the piezo. 
(Since you have seen the full frequency range accessible by the piezo, this is how far the PLL control can reach, and the laser tuning has to be brought back into this range if it thermally drifts too far. If a tap on the switch [up or down] makes the beat frequency hard to observe on the scope, track it with the sweep speed so you can see it. Using the stepper motor, make the beat as low frequency as possible, then try to lock and it should lock right up.)

Press the Manual/PID control button on the PID amp to lock the PLL again. 
See the beat frequency go to zero, the voltage will be constant near zero. 

Now vary the piezo amp offset voltage knob, and note that it doesnt make any (noticeable) difference to the mixer output or even the piezo amp output voltage, but the voltage output from the PID amp varies to compensate. 
There is a small difference in the mixer output, but its being amplified by the large PID amp gain. 

## 4. SHOT NOISE ON THE DETECTED PULSE TRAIN. 

Observe 2800MHz from the amplified photodiode (right before the mixer) using the signal source analyzer. 

Vary the optical attenuator by rotating the screw adjust with a small flat screwdriver. 
This is a cheap, flaky thing, and is hard to maintain at a stable loss value, but it can be done. 
(The mechanism simply interrupts a beam of light with the screw.) 

Set up the analyzer for single measurements, and it will continuously monitor the RF power and frequency. 

Choose values of RF power at least 3dB apart, and take an L(f) measurement at each. 
The analyzer can be kind of flaky too, so remeasure if the high frequencies have too much noise. 
A good measurement will have smooth traces for the high frequencies. 

Note the noise floor at high frequency as the optical power is decreased, compare with short pulse limit shot noise from the lecture. 
High frequency features that appear at high power like rocks at low tide may be AM-to-PM coming from intensity noise. 

## 5. LOCKED VERSUS UNLOCKED NOISE. 

With the optical power turned back up (screw on attenuator backed out), observe the amplified 2800MHz before the mixer on the signal source analyzer. 

Take L(f) measurements when the laser is locked to the reference and when it is unlocked. 

Can you see a difference in the low frequency noise? 

Where in frequency does the lock start to make a difference? 
Maybe at a few kHz. This is the bandwidth limitation of the piezo actuator in the laser cavity, so no improvement can be made at higher frequencies. 

Take an L(f) measurement of the reference clock 2800MHz. 

Compare the three measurements by plotting them on the same graph. 
The PLL should make the laser noise follow the reference clock, below the piezo bandwidth. 

## 6. BUCKET AMBIGUITY. 

The slow scope should be set to 10ns per horizontal division, displaying the mixer IF output, the 100MHz from the photodiode (after the low pass filter) and the 100MHz from the master clock (which should be set as the trigger source). 

Put the two 100MHz signals near each other so you can compare. 

Unlock and relock the loop several times (using the button on the PID amp marked manual/PID). 

See that when unlocked, the mixer output jumps up and down as the laser 100MHz rolls past the clock 100MHz. 
Each time the loop is re-locked, the relative phase of the 100MHz signals is different. 
Since this is what matters because it corresponds to the actual pulses, this is a problem if only locking to a high harmonic: you dont know which of 28 places within 10ns the pulse will wind up. 
So, 100MHz should also be transmitted, and taken into account when designing the control.
