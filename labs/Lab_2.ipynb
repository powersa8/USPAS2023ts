## 1. CALIBRATE DELAY MEASUREMENT USING MIXER. 

With the loop unlocked (PID amp button on lower right pushed such that LED next to “manual” is lit), vary the optical delay and see the IF outputs on both mixers displayed on the slow scope. 
The two voltages may vary in the same direction or opposite directions depending on the installed cable lengths. 
The mixer outputs will depend on input signal levels, but the mixer looking at the retroreflected signal should see about twice the effect because it is receiving a signal that is double-passed through the delay. 

Vary the RF delay and see only the out-of-loop mixer IF vary. 

Adjust the optical delay and the RF delay to make the mixers’ outputs both approximately zero at the same time. 

Calibrate the out-of-loop mixer voltage-versus-time factor in the same way as for experiment 1. 
That is: adjust the optical delay through a range near zero and large enough to be easily measured, but much less than half a wave. 
One can figure what sort of range is acceptable by knowing the mixer output is sinusoidal, and choosing a number of degrees consistent with the “small angle approximation” for which the sine function is linear about zero. 

Then, record the out-of-loop voltage at the ends of this optical delay range, and find the slope. 
Now you have a calibration and can read the delay by measuring IF voltage.  



## Goal: Demonstrate delay-stabilized transmission of RF over fiber. 

**Key elements:**

Laser and modulator, line length control, automatic controller. 

**What is happening:**

A CW (continuous wave) laser signal is amplitude modulated by an external, electrooptic modulator driven with RF. The modulated signal is sent to a receiver through a retroreflector, which allows some of the signal to pass, while the back-reflected signal is phase compared with the sent signal (after having transited the fiber twice). The back-reflected signal is compared in phase to the original transmitted signal by a mixer, which is set up to produce zero voltage at center, and a +/- signal on either side. This is amplified by a PID amplifier, which actuates piezoelectric fiber spools to maintain a constant phase with respect to the reference.
The long transmission fiber would in practice be inserted between the piezo spools and the retroreflector.

## 2. FIND PIEZO DELAY RANGE. 

Again, with PID unlocked, adjust the piezo amp offset voltage using the front panel knob, and observe the two mixer voltages. 

Set the piezo amp offset voltage to zero and adjust the optical delay to bring the out-of-loop mixer IF voltage back to zero. 

Adjust the piezo amp from 0 to 150V and measure the delay change using mixer voltage. 

Verify by adjusting the optical delay to bring the mixer voltage back to zero. 

To get better precision, use the Vernier scale on the distance, and remember that the optical delay is double-passed. 
Maybe an even finer measurement can be had using the crank rotation, but there is some backlash and no scale on it. Worth a try. 

Plot the mixer voltage versus piezo voltage as you go from 0 to 150 and 150 to 0. 
If you have enough points, you may be able to see hysteresis in the piezo response. I’ve not done this myself, but it should be interesting.

## 3. 

Set the piezo amp offset voltage to 75 Volts. 

Using the optical delay, go back to the condition with both mixer IF voltages at zero. 

Lock the loop by pushing the lower right button on the PID amp, to light the “PID control” LED. 

Adjust the optical delay until the PID output voltage (on the bar graph on the amp) is positive, and within 0 to 10V. 
Then, the piezo amp voltage should be 0 to 150V. 
See, the piezo amp is ”single ended” rather than plus and minus, so half of the range of the PID amp is unused. But, still, the whole piezo range is utilized. 

Vary the optical delay and watch the two mixer voltages on the scope. 
When the delay is too far for the piezos to compensate, the voltages will vary as you tune the delay. At some point, the delay is within the piezo range, and the voltages stop moving, staying near zero. As as the delay continues to change, the PID output and the piezo amp voltage will go from 0 to 10V and 0 to 150V, respectively. 

The piezo amp has gain 15. when the range is again exceeded, the voltages will resume changing, not compensated by the loop anymore. Well, that’s it, a delay controlled RF-over-fiber link. 

Record the delay range over which it locks.
